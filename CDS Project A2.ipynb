```python
hp3 = hp1["LotArea"].quantile(0.99)
```

```python
hp4 = hp1[hp1["LotArea"]<hp3] ## removing outliers
```

```python
X = pd.DataFrame(hp4, columns=['LotArea', 'TotalBsmtSF', '1stFlrSF','2ndFlrSF'])
fig = plt.figure()
ax1 = fig.add_subplot(111)
x = X['LotArea']
y = X['TotalBsmtSF']
y2 = X['1stFlrSF']
y3 = X['2ndFlrSF']
#ax1.scatter(X['LotArea'], X['TotalBsmtSF'])
#ax1.scatter(X['LotArea'], X['1stFlrSF'])
ax1.plot(np.unique(x), np.poly1d(np.polyfit(x,y2,1))(np.unique(x)), label = "1st Floor Square Feet")
ax1.plot(np.unique(x), np.poly1d(np.polyfit(x,y,1))(np.unique(x)),label ="Total Basement Square Feet")
ax1.plot(np.unique(x), np.poly1d(np.polyfit(x,y3,1))(np.unique(x)), label = "2nd Floor Square Feet")
#ax1.scatter(X['LotArea'], X['2ndFlrSF'])
plt.title('Various House Areas vs. Lot Area')
plt.xlabel('LotArea')
plt.ylabel('Square Feet')
plt.legend()
plt.show()
```
